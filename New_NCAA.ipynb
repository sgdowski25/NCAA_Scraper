```python
import requests, re
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import warnings, time
from time import sleep
warnings.filterwarnings('ignore')
url = "https://www.sports-reference.com/cbb/postseason/"
url2= "https://www.sports-reference.com/"
page = requests.get(url, verify=False)
soup = BeautifulSoup(page.text,'html.parser')
tournaments = soup.find_all('th')
brackets = soup.find_all('a', href=re.compile(r"^/cbb/postseason/"))
linklist = [urljoin(url, a['href'])for a in brackets]
del linklist[0]
del linklist[82:88]
```

```python
yearurl="https://www.sports-reference.com/cbb/postseason/2011-ncaa.html"
yearpage = requests.get(yearurl, verify=False)
yearsoup = BeautifulSoup(yearpage.text,'html.parser')
games = yearsoup.find_all('a', href=re.compile(r"^/cbb/boxscores/\d+"))
gamelist = [urljoin(url, a['href'])for a in games]
del gamelist[196:200]
```

```python
gameurl="https://www.sports-reference.com/cbb/boxscores/2015-03-19-lafayette.html"
gamepage = requests.get(gameurl, verify=False)
gamesoup = BeautifulSoup(gamepage.text,'html.parser')
```

```python
sleep(1.5) #space out your requests so you don't get blocked from the server!
all_stats=[]
labeldata= []
data=[]
statlabels= gamesoup.find_all('thead')
for row in statlabels:
    label = row.find_all('th')
    for n, l in enumerate(label):
        if n>2:  
            labeldata = [n.text.replace('\n','').strip() for n in label]
            del labeldata[:3]
            print(labeldata)
    basic_stats= gamesoup.find_all('tfoot')
    for row in basic_stats:
        stat = row.find_all('td')
        data = [n.text.replace('\n','').strip() for n in stat]
        print(data)
team_stats= zip(labeldata,data)
#print(list(team_stats))
```
